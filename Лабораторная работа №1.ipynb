### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
def sort_selection(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] > min(arr[j:]):
                arr[i], arr[j] = arr[j], arr[i]
    return arr
sel_matrix = copy.deepcopy(matrix)
start_time = time.time()
# for line in sel_matrix:
#     sort_selection(line)
end_time_selection = (time.time() - start_time)*1000

print(f"Сортировка выбором: {end_time_selection} ms", )
```

```python
# Сортировка вставкой.
def sort_insertion(arr):
    for i in range(1, len(arr)):
        for j in range(i, 0, -1):
            if arr[j] < arr[j - 1]:
                arr[j - 1], arr[j] = arr[j], arr[j - 1]
            else: break
    return arr

for line in matrix:
    sort_insertion(line.copy())

start_time = time.time()
# for line in matrix:
#     print(sort_insertion(line.copy()))
end_time_invertion = (time.time() - start_time)*1000
print(f"Сортировка вставкой: {end_time_invertion} ms", )
```

```python
# Сортировка обменом. || Сортировка пузырьком.
def sort_bubble(arr):
    for i in range(0, len(arr) - 1):
        for j in range(0, len(arr) - 1 - i):
            if arr[j + 1] < arr[j]:
                arr[j + 1], arr[j] = arr[j], arr[j + 1]

    return(arr)

start_time = time.time()
# for line in matrix:
#     print(sort_bubble(line.copy()))
end_time_bubble = (time.time() - start_time)*1000
print(f"Сортировка пузырьком: {end_time_bubble} ms", )    
```

```python
# Сортировка Шелла.
def sort_Shell(arr : list):
    rangeValue = len(arr) // 2
    while rangeValue > 0:
        for i in range(rangeValue, len(arr)):
            currentValue = arr[i]
            index = i
            while index >= rangeValue and arr[index - rangeValue] > currentValue:
                arr[index] = arr[index - rangeValue]
                index -= rangeValue
                arr[index] = currentValue
        rangeValue //= 2
    return arr 
start_time = time.time()
# for line in matrix:
#     print(sort_Shell(line.copy()))
end_time_Shell = (time.time() - start_time)*1000
print(f"Сортировка Шелла: {end_time_Shell} ms", )    
```

```python
# Быстрая сортировка.
def sort_quick(arr):
    if len(arr) <= 1:
        return arr
    mainElem = arr[0]
    left = [i for i in arr if i < mainElem]
    center = [i for i in arr if i == mainElem]
    right = [i for i in arr if i > mainElem]

    return sort_quick(left) + center + sort_quick(right)
start_time = time.time()
# for line in matrix:
#     print(sort_quick(line.copy()))
end_time_quick = (time.time() - start_time)*1000

print(f"Быстрая сортировка: {end_time_quick} ms", )     
```

```python
# Турнирная сортировка.
def tournament_sort(arr):
    n = len(arr) 
    tree = [None] * (2 * n - 1) # создаем список для хранения дерева турнира

    def build_tree(node, start, end): # функция для построения дерева
        if start == end: 
            tree[node] = arr[start]
        else: # иначе делим наш диапазон пополам и рекурсивно строим левое и правое поддерево
            mid = (start + end) // 2
            build_tree(2 * node + 1, start, mid)
            build_tree(2 * node + 2, mid + 1, end)
            tree[node] = max(tree[2 * node + 1], tree[2 * node + 2]) # записываем в текущий узел максимальный элемент из левого и правого поддерева

    def query_tree(node, start, end, left, right): # функция для поиска максимального элемента на заданном диапазоне
        if left > end or right < start: 
            return -float('inf')
        if left <= start and right >= end: 
            return tree[node]
        mid = (start + end) // 2 
        return max(query_tree(2 * node + 1, start, mid, left, right), 
                   query_tree(2 * node + 2, mid + 1, end, left, right))

    def find_winner(l, r): # функция для определения победителя между двумя индексами массива
        return l if arr[l] < arr[r] else r

    def tournament(start, end): # функция для проведения турнира
        if start == end: 
            return start
        mid = (start + end) // 2 
        left = tournament(start, mid)
        right = tournament(mid + 1, end)
        return find_winner(left, right)
    build_tree(0, 0, n - 1)
    sorted_arr = []
    for i in range(n):
        winner_index = tournament(0, n - 1 - i)
        sorted_arr.append(arr[winner_index])
        arr.pop(winner_index)
    return sorted_arr
start_time = time.time()
# for line in matrix:
#     print(tournament_sort(line.copy()))
end_time_tour = (time.time() - start_time)*1000

print(f"Турнирная сортировка: {end_time_tour} ms", )
```
